### Conditional Statement (if , else , elif)

if statement is a control flow statement that allows you to execute a block of code only if a specified condition is true.

```python
## if statement 
age = 20
if age >= 18:
  print("You are eligible to vote.")
```

else statement is a block of code that will be executed if the condition in the if statement is False.

```python
# else statement 

age = 16
if age >= 18:
  print(" You are eligible to vote.") 
else : 
  print(" You are not eligible to vote.")
```

elif statement is used to check multiple conditions

```python
# elif statement 
age = int(input(" enter your age : "))

if age < 12 : 
  print(" you are a child.")
elif age <18 :
  print(" you are a teenager. ")
else : 
  print(" you are a adult . ")
```

Nested Conditional Statements . you can place  if , elif , else statements inside another if , elif , else statements.

```python
#Nested Conditional Statements 

num = int(input("enter a number : "))
if num >= 0:
  print(" The number is positive.")
  if num%2 == 0 : 
    print(" The number is even .")
  else :
    print(" The number is odd . ")
else : 
  print(" The number is negative .")    
```

```python
# Practical example  
# determine if a year is a leap year using nested conditional statements 
year = int(input("enter a year : "))
if (year % 4) == 0:
    if (year % 100) == 0:
        if (year % 400) == 0:
            print(f"{year} is a leap year.")
        else:
            print(f"{year} is not a leap year.")
    else:
        print(f"{year} is a leap year.")
else :
    print(f"{year} is not a leap year.")        

```
