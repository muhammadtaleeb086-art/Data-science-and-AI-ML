# Loops

### For Loop
 For loops is a control flow statement that allows code to be executed repeatedly based on a sequence or iterable object. It is commonly used to iterate over lists, tuples, strings, and other iterable types in Python.

```python
# for loops 
for i in range (5):
    print("Iteration:", i) # iteration means repetition of a process
```

```python
for i in range(1,6):
  print(i)
```

```python
for i in range (1,10,1):# step count
    print(i)
```

```python
for i in range (1,10,2):# step count 
  print(i)
```

```python
for i in range(10, 1 , -1): # decrementing order
    print(i)
```

```python
for i in range (10,1 , -2):
  print(i)
```

```python
# strings in for loops

str = "Taleen Chouhan"

for i in str:
  print(i)
```

### While Loop 
While loops is a control flow statement that allows code to be executed repeatedly based on a given Boolean condition.

```python
# while loops 
# the while loop continues to execute a block of code as long as a condition is true.

count = 0 
while count<5:
  print("count is : ", count)
  count +=1  # count = count + 1
```

```python
count = 0
while count%2==0:
  print("Even number:", count)
  count += 1 # count = count + 1
```

# Loop Control Statements 
loop control statements is used to change execution from its normal sequence and it includes break , continue and pass statements.

### Break statement 
break statement is used to terminate the loop prematurely when a certain condition is met.

```python
# break statement 
for i in range (10):
  if i == 5 :
    break
  print(i)
```

### Continue statement 
Continue statement is used to skip the current iteration of a loop and move to the next iteration.

```python
#continue statement 
for i in range (10):
  if i == 5 :
    continue
  print(i)
```

```python
for i in range (10):
  if i % 2 == 0: # even number
    continue
  print(i)
```

### Pass statement 
pass statement is a null operation . when it is executed, nothing happens. It is used as a placeholder.


```python
#pass statement 
for i in range (5):
  if i == 3:
    pass
  print(i)
```

## Nested loop
nested loops is a loop inside a loop

```python
# nested loops 
for i in range (3):
  for j in range (2):
    print(f"i : {i} and j : {j}")
```

```python
# example 
# calcuate the sum of frist N natural number using a while loop and for loop

# while loop 
number = 10
sum = 0
count = 1

while count <= number :
  sum = sum  + count 
  count = count + 1
print(" the sum of first ", number , " natural number is : ", sum)
```

```python
# for loop 
number = 10
sum = 0
for i in range (11):
  sum = sum + i
print(" the sum of first ", number , " natural number is : ", sum)
```

```python
# example  prime number between 1 to 100

for num in range (1 , 101 ):
  if num > 1:
    for i in range (2 , num):# check for factors
      if num % i == 0: # not a prime number
        break 
    else :
      print(num)
```
