Syntax refers to the set of rules that define the combination of symbols that are considered to be correctly structured program in a language . In simple terms , syntax is about the correct arrangement of words and symbols in a code.


Semantics refers to the meaning or the interpretation of symbols, characters, and commands in a language . it is about what ther code is supposed to do when it run.

```python
## basic syntax rules in python
## Case Sensitivity
# Python is case-sensitive, meaning that variables, functions, and other identifiers must be used with the exact same capitalization as they were defined.

name = "Taleeb"
Name = "Afreen"
print(name)  # Output: Taleeb
print(Name)  # Output: Afreen 
```

```python
## Indentation
# Python uses indentation (whitespace at the beginning of a line) to define the structure and scope of code blocks, such as functions, loops, and conditionals. Proper indentation is crucial for code readability and functionality. 

age = 18
if age >= 18:
    print("You are an adult.") # Correct indentation

print(age)
```

```python
#Line Continuation
# In Python, if a statement is too long to fit on a single line, you can use a backslash (\) to indicate that the statement continues on the next line.
total = 1 + 2 + 3 + \
        4 + 5 + 6
print(total)  # Output: 21
```

```python
##multiple statements on a single line
# You can write multiple statements on a single line by separating them with a semicolon (;). However, it's generally recommended to use separate lines for better readability.
x = 5; y = 10; z = x + y
print(z)  # Output: 15
```

```python
##Understaanding variable types in Python
#Python is a dyanamically typed language which means you don't need to declare variable types explicitly. The interpreter infers the type based on the assigned value.
name = "Taleeb"  # String
age = 21          # Integer
height = 5.7     # Float
is_student = True # Boolean
print(type(name))      # Output: <class 'str'>
print(type(age))       # Output: <class 'int'>
print(type(height))    # Output: <class 'float'>
print(type(is_student)) # Output: <class 'bool'>
```

```python
# Type Conversion
# You can convert between different variable types using built-in functions.
Age = 21          # Integer
print(type(Age))       # Output: <class 'int'>
age_str = str(Age)    # Convert to String
print(type(age_str))   # Output: <class 'str'>
```

```python
age="21"  # String
print(type(age))       # Output: <class 'str'>
print(type(int(age)))   # Convert to Integer
```

```python
#Dynamic Typing
# python allow the type of variable to change as the program executes.

var = 10  # Initially an integer
print(var, type(var))

var = "taleeb"  # Now a string
print(var,type(var))

var = 5.7  # Now a float
print(var, type(var))
```
