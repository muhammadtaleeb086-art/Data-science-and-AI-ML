Datatypes
 
1.Defination of Datatype

 Datatype is a classification of data which tells the compiler or interpreter how the programmer intends to use the data.

 They determine the type of operations that can be performed on the data, the value  that the data can take , and the amount of memory that needed to store  the data.

2. Importance of Data types in programming 

 2.1: Datatypes ensure that data is stored in an efficient way.

 2.2: they help in perfoming correct operations on data.

 2.3: Proper use of data types can prevent errrors and bugs in the problems. 


```python
#interger example 
age = 21 
print(age)
print(type(age))
```

```python
#floating point example
height = 5.7
print(height)
print(type(height))
```

```python
#string datatype example 
name = "taleeb"
print(name)
print(type(name))
```

```python
#boolean datatype example 
I_am_engineer = True
print(I_am_engineer)
print(type(I_am_engineer))
```

```python
#comman error 
result = "hello " + 5
print(result)
```

```python
#correct way 
result = "hello" + str(5) # type casting
print(result)
```

Operators 

 Operators is a special symbol that performs specific operations on one, two, or more operands and produces a result.

### Arthmetic Operations

 Addition,
 Subtraction,
 Multiplication,
 Division,
 Floor Division,
 Modulus.

```python
#Arthmetic Operators
a = 10 
b = 5
print("Addition: ", a +b)
print("Subtraction :", a-b)
print("Multiplication :", a*b)
print("Division :", a/b)
print("Floor Division :", a//b) # removes decimal part
print("Modulus :", a%b) # remainder
print("Exponent :", a**b) # a raised to the power b
```

### Comparison Operators

```python
#comparison Operators
x = 10
y = 20
print("Equal to :", x == y)
print("Not Equal to :", x != y)
print("Greater than :", x > y)
print("Less than :", x < y)
print("Greater than or Equal to :", x >= y)
print("Less than or Equal to :", x <= y)
```

### Logical Operators

```python
#Logical Operators
# And operator
# True if both operands are true
p = True
q = False
print("Logical AND :", p and q)
```

```python
# OR operator
# True if at least one operand is true
p = True
q = False
print("Logical OR :", p or q)
```

```python
# Not operator
# True if operand is false
p = True
print("Logical NOT :", not p)
```
