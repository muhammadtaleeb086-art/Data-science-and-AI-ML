# Sets

Sets are a built-in data type in python used to store collection of unique items .They are unordered , meaning that the elements do not follow a specific order , and they do not allow duplicate elements . Sets are useful for membership tests , eleminating duplicates entires and perfoming mathematical set operation like union , intersection , difference and symmertric difference .

```python
# create a set 
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

## Set Operations in Python 

1. Union
Combines all unique elements from two or more sets.
Duplicate values are removed automatically.

2. Intersection
Returns only the elements that are common in all given sets.

3. Difference
Returns elements that exist in one set but not in the other.
The order of sets matters here.

4. Symmetric Difference
Returns elements that are present in either set but not in both.

5. Subset
Checks whether all elements of one set are present in another set.

6. Superset
Checks whether a set contains all elements of another set.

7. Disjoint
Checks whether two sets have no common elements at all.

Important Points About Sets

Sets store only unique elements

Sets are unordered and unindexed

Sets are mutable, but their elements must be immutable

Set operations are mainly used for comparison and filtering data

```python
## Basics sets operation 

## adding and removing elements 

my_set = {1,2,3,4,5,6,7,}
my_set.add(10)
print(my_set)
my_set.add(10)
print(my_set)
```

```python
#remove the elements from a set 
my_set = {1,2,3,4,5,6,7,10}
my_set.remove(10)
print(my_set)
my_set.discard(11)
print(my_set)
```

```python
## pop method - FIFO method 
remove_element = my_set.pop()
print(remove_element) # first in fisrt out 
print(my_set)
```

```python
## clear all the elements 
my_set.clear()
print(my_set)
```

```python
## set membership test 
my_set = {1,2,3,4,5,6,7}
print(3 in my_set) # check in 3 agar 3 hai to true otherwise false
print(10 in my_set)
```

```python
#mathematical operation 

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## Union method = mean combining all sets 
union_set = set1.union(set2)
print(f"Union Set : {union_set}")

## Intersection method : select common things 
intersection_set =set1.intersection(set2)
print(f"Intersection Set : {intersection_set}")


set1.intersection_update(set2) # set 1 ko update kr k intersection elements ko set1 kr degya 
print(f"Set 1 intersection update : {set1}")
```

```python


## Diffrence  = jis set ky difffrence dehk reh hai uske unique value lega sirf comman values remove kr degya 
print(set1.difference(set2))
print(set2.difference(set1))
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}
```

```python
## symmetric difference  == alg alg set ke unique elements ko combine kr degya .. 

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}
print(set1.symmetric_difference(set2))
```

```python
## Sets methods
set1 = {1,2,3,4,5}
set2 = {3,4,5}

## is subset  = checking ki subset hai ya nhi 
print(set1.issubset(set2))

## is super set = checking ki super set hai ya nhi 
print(set1.issuperset(set2))
```

```python
# task ki dublicates ko remove kre list mai se 

lst = [1,2,2,3,4,5,5,6,7,8,8]

print(set(lst))
```

```python
## counting unique words in text 

text = " My name is Muhammed Taleeb Chouhan"
words = text.split()

## convert list of words to set to get unique words 

unique_words = set(words)
print(unique_words)
print(len(unique_words)) 
```

## Conclusion on set in Python


In Python, a set is a powerful data structure that allows us to store unique, unordered elements. It is especially useful when we need to eliminate duplicates, perform mathematical operations like union, intersection, and difference, or check membership efficiently. Sets are highly optimized for such operations, making them faster compared to lists in many scenarios.
